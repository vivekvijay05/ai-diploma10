## Matplotlib

Matplotlib is an easy-to-use, low-level data visualization library that is built on NumPy arrays. It provides a lot of flexibility.

    Scatter Plot  - plt.scatter()
    Line Chart    - plt.plot()
    Bar Chart	  - plt.bar()
    Histogram	  - plt.hist()
    pie chart - plt.pie()
    boxplot  -plt.boxplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
display(df)
```

```python
df.describe()
```

```python
# Scatter plot with day against tip


plt.scatter(df['total_bill'], df['tip'])# visualizing relationships between two numerical variables.

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('total_bill')
plt.ylabel('Tip')

plt.show()
```

demo practice

```python
plt.scatter(df["smoker"],df["time"])

plt.title("SCATTER PLOT OF SMOKER AND TIME")

plt.xlabel("smoker")
plt.ylabel("time")

plt.show()
```

```python

# line plot with day against tip


plt.plot(df['size'])  # represent the relation between two data X and Y on a different axis
plt.plot(df['tip'])

# Adding Title to the Plot
plt.title("line Plot")


plt.show()
```

demo practice

```python
plt.plot(df["day"])
plt.plot(df["tip"])

plt.title("LINE PLOT OF THE DATA")
plt.show()
```

```python
# Bar chart with day against tip


plt.bar(df['day'], df['tip'])#to visualize categorical data

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

# Adding the legends
plt.show()
```

demo practice

```python
plt.bar(df["total_bill"],df["day"])

plt.title("BAR CHART IF DAYS")

plt.xlabel("total_bill")
plt.ylabel("day")

plt.show()
```

```python
# histogram of total_bills


plt.hist(df['total_bill'])# graphical representation of data distribution

plt.title("Histogram")

# Adding the legends
plt.show()
```

demo practice

```python
plt.hist(df["sex"])

plt.title("HISTOGRAM OF GENDER")
plt.show()
```

```python
plt.pie(df['tip'],labels=df['day'])# visual representation of data that makes it easy to compare parts of a whole or represent numerical values
plt.show()
```

```python
plt.boxplot(df["total_bill"])# display the summary of a data set, including minimum, first quartile, median, third quartile and maximum
plt.show()
```

## Seaborn

    Scatter Plot  - sns.scatterplot()
    (Easy use of the hue argument for coloring by category (e.g., sex))

    Line Plot - sns.lineplot()
    (Simplifies creating line charts, even without specifying X and Y)

    Bar Plot  - sns.barplot()
    (Easily incorporates categorical variables for comparison)

    Histogram - sns.histplot()
    (Includes an option for a Kernel Density Estimate (kde=True))


```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df=sns.load_dataset('tips')
display(df)

```

```python
# scatter plot

#plt.scatter(df['day'],df['tip'])#matplot
sns.scatterplot(x='day', y='total_bill', data=df)
# setting the title using Matplotlib
plt.title('Day - Tip')
plt.show()
```

demo practice

```python
sns.scatterplot(x="tip",y="day",data=df)
plt.title("tip day graph")
plt.show()
```

```python
sns.regplot(x= "tip",y="total_bill",data=df) #  to represent the regression lines
plt.title("regression lines graph")
plt.show()
```

```python
plt.plot(df['time'],df['tip'])
plt.show()
```

```python
# draw lineplot



sns.lineplot(x="day", y="tip", data=df) #

# setting the title using Matplotlib
plt.title('time - tip')

plt.show()
```

```python
plt.boxplot(df['tip'])
plt.show()
```

```python
plt.bar(df['day'],df['tip'])#matplot
plt.show()
```

```python
# barplot
sns.barplot(x='day',y='tip', data=df, hue='time')

plt.show()
```

```python
# histplot

sns.histplot(x='total_bill', data=df, kde=True,hue="time")

plt.show()


```

skew - distribution

    Right skewed - data-> left - mean>median>mode
    left skewed - data-> right - mode>median>mean

Kurtosis - peakedness

    leptokurtic - high
    mesokurtic - normal
    platykurtic - flat

```python

plt.figure(figsize=(8, 5))
sns.kdeplot(df, fill=True, color="skyblue", linewidth=2)

# Add labels and title
plt.title("Kernel Density Estimate (KDE) Plot", fontsize=14)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Show the plot
plt.show()
```
