# Central tendencies for imputation
1. mean
2. median
3. mode

```python
#analyze the distribution of data in Salary
```

```python

sns.distplot(data.salary)
```

```python
data['salary'].value_counts()
```

```python
#Replace the missing values with median

data['salary'].fillna(data['salary'].mode()[0],inplace=True)
display(data)
```

# drop the values

```python
data['salary'].dropna(inplace=True)
```

drop duplicated values

```python
data.drop_duplicates()
```
