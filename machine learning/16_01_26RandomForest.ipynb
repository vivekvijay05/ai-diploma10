```python
feature=df.drop(columns="DEATH_EVENT",axis=1)
target=df["DEATH_EVENT"]
```

```python
feature
```

```python
df["DEATH_EVENT"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
feature_train,feature_test,target_train,target_test=train_test_split(feature,target,test_size=0.3,random_state=32)
```

```python
feature_train
```

```python
target_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=80,max_depth=4,bootstrap=True,max_samples=100)
model.fit(feature_train,target_train)
```

```python
predictions=model.predict(feature_test)
prediction_df=pd.DataFrame({'Actual': target_test, 'Predicted': predictions})
prediction_df
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(target_test,predictions))
print(classification_report(target_test,predictions))
```
