```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Employment"]=le.fit_transform(df["Employment"])
df["MaritalStatus"]=le.fit_transform(df["MaritalStatus"])
```

```python
feature=df.drop(columns=["Loan"],axis=1)
target=df["Loan"]
```

```python
from sklearn.model_selection import train_test_split
feature_train,feature_test,target_train,target_test=train_test_split(feature,target,test_size=0.2,random_state=42)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
feature_train=sc.fit_transform(feature_train)
feature_test
```

```python
feature_test=sc.fit_transform(feature_test)
feature_test
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=4,min_samples_split=5)
model.fit(feature_train,target_train)
```

```python
predic=model.predict(feature_test)
predic=pd.DataFrame(predic,columns=["Predicted"],index=target_test.index)
predic
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy=accuracy_score(target_test,predic)
print(accuracy)

print(classification_report(target_test,predic))
```

```python
predic.value_counts()
```

```python
merged=pd.concat((target_test,predic),axis=1)
merged
```

```python
merged_sorted = merged.sort_values(by=['Loan', 'Predicted']).reset_index(drop=True)  #with sorted prediction just for knowledge
display(merged_sorted)
```
