#Sentiment Analysis using TextBlob and VADER

Sentiment Analysis is a Natural Language Processing (NLP) technique used to determine the emotional tone or opinion expressed in text.

It is widely used to analyze opinions in:

- product reviews

- social media posts

- customer feedback

- news articles

    Sentence:

       I love this product

    Sentiment:

      Positive


###Types of Sentiment

| Sentiment | Example                  |
| --------- | ------------------------ |
| Positive  | I really like this movie |
| Negative  | The service was terrible |
| Neutral   | The meeting is at 10 AM  |


## Sentiment Analysis Techniques

There are different approaches to sentiment analysis:

In this topic we focus on rule-based tools:

- TextBlob

## Sentiment Analysis using TextBlob

###What is TextBlob?

TextBlob is a Python library built on top of NLTK that provides simple APIs for common NLP tasks.

Features:

- sentiment analysis

- noun phrase extraction

- part-of-speech tagging

- translation

```python
!pip install textblob
```

```python
!python -m textblob.download_corpora
```

```python
from textblob import TextBlob

text = "I love AI"

analysis = TextBlob(text)

print(analysis.sentiment)
```

###TextBlob Sentiment Scores


| Metric       | Range    | Meaning                 |
| ------------ | -------- | ----------------------- |
| Polarity     | -1 to +1 | Negative to Positive    |
| Subjectivity | 0 to 1   | Objective to Subjective |

| Sentence            | Polarity | Sentiment |
| ------------------- | -------- | --------- |
| I love this movie   | 0.8      | Positive  |
| I hate this product | -0.7     | Negative  |
| This is a table     | 0        | Neutral   |


## Sentiment Analysis using VADER
###What is VADER?

VADER (Valence Aware Dictionary and sEntiment Reasoner) is a rule-based sentiment analysis tool specifically designed for social media text.

It is included in the NLTK library.

VADER handles:

- emojis

- slang

- punctuation

- capitalization

Example:

    This movie was AMAZING!!!

    VADER detects strong positive sentiment.

```python
!pip install nltk
import nltk
nltk.download('vader_lexicon')

#
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

text = "I love this phone"

score = sia.polarity_scores(text)

print(score)
```

| Score    | Meaning                 |
| -------- | ----------------------- |
| pos      | Positive sentiment      |
| neu      | Neutral sentiment       |
| neg      | Negative sentiment      |
| compound | Overall sentiment score |


##Applications of Sentiment Analysis

Sentiment analysis is used in many industries.

- Social Media Monitoring : Companies analyze customer opinions on social media.

- Product Reviews : E-commerce websites analyze reviews to understand customer satisfaction.

        Example:
        Amazon reviews
        Flipkart reviews

- Market Research : Businesses analyze public opinion about products or services.

- Political Analysis : Analyze public sentiment about political leaders and policies.

- Customer Support : Companies analyze customer feedback automatically.

```python
from textblob import TextBlob
from nltk.sentiment import SentimentIntensityAnalyzer

text = "The movie was worst!"

# TextBlob
blob = TextBlob(text)
print("TextBlob polarity:", blob.sentiment.polarity)

# VADER
sia = SentimentIntensityAnalyzer()
print("VADER:", sia.polarity_scores(text))
```
