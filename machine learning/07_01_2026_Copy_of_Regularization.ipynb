```python
import pandas as pd
```

**Read the housing dataset file**

```python
data = pd.read_csv('/content/Housing_Dataset.csv')
data.head()
```

```python
data.shape
```

**Number of nan values in the given dataset**

```python
data.isnull().sum()
```

```python
data.info()
```

```python
usefulCols = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'Price']

newData = data[usefulCols]
```

```python
newData.head()
```

```python
newData.shape
```

```python
newData.isnull().sum()
```

```python
cols_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
newData[cols_fill_zero] = newData[cols_fill_zero].fillna(0)
```

```python
newData.isnull().sum()
```

```python
newData['Landsize']=newData['Landsize'].fillna(newData['Landsize'].median())
newData['BuildingArea'] = newData['BuildingArea'].fillna(newData['BuildingArea'].median())
```

```python
newData.isnull().sum()
```

```python
newData.dropna(inplace=True)
newData.isnull().sum()
```

```python
newData.head()
```

```python
newData['Suburb'].value_counts()
```

```python
newData.shape
```

```python
newData=pd.get_dummies(newData,drop_first=True)
```

```python
newData.shape
```

```python
## X - input features
## y - target feature
X = newData.drop(columns='Price', axis=1)
y = newData['Price']
```

**Splitting the data into train and test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=2)
```

**Data Modeling - Linear Regression**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

**Problem of Overfitting**

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

```python
model.coef_
model.intercept_
```

**Regularization Method : Lasso Regularization(L1)**

```python
from sklearn.linear_model import Lasso
lassoModel = Lasso(alpha=100, max_iter=1000, tol=.1)
lassoModel.fit(X_train, y_train)
```

```python
lassoModel.score(X_test, y_test)
```

```python
lassoModel.score(X_train, y_train)
```

```python
lassoModel.coef_
```

```python

zero_coeff_features = X_train.columns[lassoModel.coef_ == 0]

print("Features with coefficients converted to 0 by Lasso Regularization:")
for feature in zero_coeff_features:
    print(feature,end="  | ")
```

```python
#parameter tuning
from sklearn.linear_model import LassoCV

lasso_cv = LassoCV(alphas=[10, 20, 50, 100, 200],
                   cv=5, max_iter=10000)

lasso_cv.fit(X_train, y_train)
```

```python
lasso_cv.alpha_
```

```python
from sklearn.linear_model import Lasso

best_lasso = Lasso(alpha=lasso_cv.alpha_)
best_lasso.fit(X_train, y_train)


```

```python
best_lasso.score(X_test, y_test)
```

```python
best_lasso.score(X_train, y_train)
```

**Regularization Method: Ridge Regularization(L2)**

```python
from sklearn.linear_model import Ridge
ridgeModel = Ridge(alpha=50, max_iter=100, tol=0.1)
ridgeModel.fit(X_train, y_train)
```

```python
ridgeModel.score(X_test, y_test)
```

```python
ridgeModel.score(X_train, y_train)
```

```python
ridgeModel.coef_
```

```python
from sklearn.linear_model import RidgeCV

ridge_cv = RidgeCV(alphas=[0.1, 20, 50, 100, 200], cv=5, scoring='r2')

ridge_cv.fit(X_train, y_train)
```

```python
ridge_cv.score(X_test, y_test)
```

```python
ridge_cv.score(X_train, y_train)
```

## ElasticNet (L1 + L2)

```python
from sklearn.linear_model import ElasticNetCV

enet_cv = ElasticNetCV(
    alphas=[0.1, 1, 10, 50, 100],
    l1_ratio=[0.2, 0.5, 0.8],
    cv=5,
    max_iter=5000
)

enet_cv.fit(X_train, y_train)

enet_cv.alpha_, enet_cv.l1_ratio_
```

```python
from sklearn.linear_model import ElasticNet
best_enet = ElasticNet(
    alpha=enet_cv.alpha_,
    l1_ratio=enet_cv.l1_ratio_,
    max_iter=5000
)

best_enet.fit(X_train, y_train)


```

```python
best_enet.score(X_test, y_test)
```

```python
best_enet.score(X_train, y_train)
```
