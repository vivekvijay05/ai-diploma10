# Word Embeddings

- Text vectorizer alone can't capture semantic meaning of the text

        eg:
            king
            queen
            man
            woman

            these words will treat as unrelated words

## Word Embeddings

- Word embeddings represent words as dense vectors.
- Each word is represented as a real-valued vector
- Words with similar meanings have similar vector representations
- Semantic relationships between words can be captured mathematically





   | Word  | Vector Representation (simplified) |
| ----- | ---------------------------------- |
| king  | [0.72, 0.15, 0.64, ...]            |
| queen | [0.70, 0.18, 0.66, ...]            |
| man   | [0.52, 0.10, 0.60, ...]            |
| woman | [0.50, 0.14, 0.62, ...]            |


These vectors are learned automatically from large text datasets.

### Why Word Embeddings are Important

Word embeddings allow machines to understand semantic similarity.

      King − Man + Woman ≈ Queen

| Word   | Similar Words       |
| ------ | ------------------- |
| doctor | physician, surgeon  |
| car    | vehicle, automobile |
| happy  | joyful, cheerful    |


###Dense vs Sparse Representation

Sparse Representation (BoW)


eg:

    [cat, dog, apple, car]

    Word "cat": [1, 0, 0, 0]
    Word "dog": [0, 1, 0, 0]

Problems:

- Mostly zeros

- No relationship between words

Dense Representation (Embeddings)

| Word | Vector             |
| ---- | ------------------ |
| cat  | [0.32, 0.45, 0.21] |
| dog  | [0.34, 0.44, 0.19] |

- Vectors are dense
- Similar words have similar vectors

### Distributional Hypothesis

Word embeddings are based on the Distributional Hypothesis:

    “Words that appear in similar contexts tend to have similar meanings.”


  eg:
     
    python is a programming language
    java is a programming language

  from the context:

    model learns that
    python ≈ java

##Types of Word Embedding Models

1. Word2Vec

- Developed by Google (2013).

Two architectures:

    - CBOW (Continuous Bag of Words)
    - Skip-Gram

2. GloVe

- Developed by Stanford

      Uses global word co-occurrence statistics.

3. FastText

- Developed by Facebook

Handles subword information, useful for:

    - rare words

    - morphologically rich languages
