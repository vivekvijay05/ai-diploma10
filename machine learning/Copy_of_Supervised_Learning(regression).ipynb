
# Supervised Learning

Supervised learning is a machine learning technique where the model learns from labeled data

    It tries to find a mapping function from X → Y such that it can predict Y for new unseen X

| Input (X)           | Output (Y)      |
| ------------------- | --------------- |
| Hours studied       | Exam score      |
| Years of experience | Salary          |
| Email text          | Spam / Not spam |


## Supervised Learning Process

    --> Collect data (with input and output)

    --> Preprocess data (handle missing values, encoding, scaling)

    --> Split data into train and test

    --> Train model on training data

    --> Test on new data

    --> Evaluate model performance

    Data (X, Y)
      ↓
    Split → Train set | Test set
      ↓
    Train → Model learns patterns
      ↓
    Test → Model predicts new Y
      ↓
    Evaluate → Accuracy, Error, etc.


## Types of Supervised Learning

| Type               | Output Type | Example                  | Algorithm                               |
| ------------------ | ----------- | ------------------------ | --------------------------------------- |
| **Regression**     | Continuous  | Predicting house prices  | Linear Regression                       |
| **Classification** | Categorical | Spam/Not Spam, Pass/Fail | Logistic Regression, Decision Tree, SVM |

Equation of Linear Regression

The mathematical equation of a straight line:


    y=mx+b

Where:

    y → Predicted value

    x → Input (independent variable)

    m → Slope (how much y changes when x changes)

    b → Intercept (value of y when x = 0)

| Metric       | Formula            | Description                      |   |                        |
| ------------ | ------------------ | -------------------------------- | - | ---------------------- |
| **MAE**      | (1/n) Σ            | yᵢ − ŷᵢ                          |   | Average absolute error |
| **MSE**      | (1/n) Σ (yᵢ − ŷᵢ)² | Penalizes large errors           |   |                        |
| **RMSE**     | √MSE               | Error in original units          |   |                        |
| **R² Score** | 1 - (RSS/TSS)      | Measures model’s goodness of fit |   |                        |

```python
import pandas as pd
df=pd.read_csv("/content/Ecommerce Customers.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

x=df.drop(columns=['Email','Address','Avatar','Yearly Amount Spent'])
y=df['Yearly Amount Spent']

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(f"x_train: {x_train.shape} \t x_test: {x_test.shape}\n y_train: {y_train.shape} \t y_test: {y_test.shape} ")
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train,y_train)

```

```python
model.coef_
```

```python
model.intercept_
```

```python
x_test
```

```python
y_test
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot scatter points in blue
sns.scatterplot(x=y_pred, y=y_test, color="blue", label="Data Points")

# Plot regression line in red (without scatter points)
sns.regplot(x=y_pred, y=y_test, scatter=False, color="red", label="Regression Line")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error
import math
```

```python
print("R2_Score = ", r2_score(y_pred, y_test))
print("MAE = ", mean_absolute_error(y_pred, y_test))
print("MSE = ", mean_squared_error(y_pred, y_test))
print("RMSE = ", math.sqrt(mean_squared_error(y_pred, y_test)))
```
