#FastText


- FastText is a word embedding technique used in Natural Language Processing (NLP) to represent words as dense numerical vectors.

- It was developed by researchers at Meta (Facebook AI Research).

- FastText improves traditional embedding models such as Word2Vec by representing words using subword information (character n-grams).

- This allows FastText to handle:

      - rare words

      - unknown words

      - morphologically rich languages

## Why FastText is Needed

- Traditional word embedding models like Word2Vec treat each word as a single unit.

Example:

      teach
      teacher
      teaching

Word2Vec treats these as three completely separate words however the root word is teach.

- FastText represents each word as a collection of character n-grams.

      Example:

      Word: learning

Character n-grams (n=3):

    <le
    lea
    ear
    arn
    rni
    nin
    ing
    ng>

The final word embedding is obtained by combining embeddings of all subwords.


##FastText Architecture


- FastText is based on the Word2Vec architecture but adds subword information.

Structure:

      Input Word
          ↓
      Character n-grams
          ↓
      Neural Network
          ↓
      Word Embedding

## FastText Training Models

FastText supports two training approaches similar to Word2Vec.

1. CBOW (Continuous Bag of Words)

Predicts a word based on its surrounding context.

    Example:

    Sentence: The cat sits on the mat

    Context words: The ___ sits

    Target word: cat

2. Skip-Gram

Predicts surrounding words using a target word.

    Example:

    Target word: cat

    Predicted context words: The, sits

| Advantage                                 | Explanation                           |
| ----------------------------------------- | ------------------------------------- |
| Handles rare words                        | Uses subword information              |
| Handles unknown words                     | Can build vectors from characters     |
| Better for morphologically rich languages | Works well with prefixes and suffixes |
| Improved accuracy                         | Captures word structure               |


### Word2Vec vs FastText

| Feature              | Word2Vec         | FastText             |
| -------------------- | ---------------- | -------------------- |
| Word representation  | Whole word       | Character n-grams    |
| Rare word handling   | Poor             | Good                 |
| Unknown words        | Cannot represent | Can generate vectors |
| Morphology awareness | No               | Yes                  |


###Using FastText in Python

FastText can be implemented using the Gensim library.

```python
pip install gensim
```

```python
from gensim.models import FastText
```

```python
sentences = [
["I","love","AI"],
["AI","is","powerful"],
["I","study","machine","learning"]
]
```

```python
model = FastText(
    sentences,
    vector_size=100,
    window=5,
    min_count=1
)
```

```python
print(model.wv["AI"])
```

```python
model.wv.most_similar("AI")
```

```python
model.wv["intelligence"] #handle unknown text
```
