# Introduction to YOLO (You Only Look Once)

- YOLO (You Only Look Once) is a real-time object detection algorithm that detects multiple objects in an image or video in a single forward pass of a neural network.



- Unlike traditional object detection models that:

       - First detect regions

       - Then classify each region

YOLO does everything in one step — making it extremely fast

- Looks at the image only once

- Divides the image into a grid

- Predicts:

      - Bounding boxes

      - Object confidence

      - Class probabilities

All at the same time.

##Working of YOLO


- Input image is divided into S × S grid cells

- Each grid cell predicts:

       - Bounding box (x, y, w, h)

       - Confidence score

       - Class probabilities

Final detections are selected using:

       - Confidence threshold

       - Non-Maximum Suppression (NMS)

###YOLO Architecture


YOLO is based on:

- CNN (Convolutional Neural Network)

- Single-stage detector

- Regression-based approach




It directly predicts:

        Class + Bounding Box Coordinates


Instead of:

        Region Proposal → Classification

### Advantages of YOLO

- Very fast (Real-time detection)
- High accuracy (especially newer versions)
- Detects multiple objects
- Works on video streams
- Easy to deploy

###Real-World Applications


- Self-driving cars
X`
- CCTV surveillance

- Face detection

- Traffic monitoring

- Medical imaging

- Industrial automation

- Drone object tracking
