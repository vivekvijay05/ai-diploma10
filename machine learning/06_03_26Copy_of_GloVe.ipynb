#GloVe (Global Vectors for Word Representation)

1. Introduction

- GloVe stands for Global Vectors for Word Representation.
- It is a popular word embedding technique used in Natural Language Processing (NLP) to represent words as dense numerical vectors.

- GloVe was developed by researchers at Stanford University in 2014.

- GloVe uses global word co-occurrence statistics from a corpus to learn word relationships.

The main idea of GloVe is:

- Words that appear frequently together in a corpus are likely to have related meanings.

| Word 1 | Word 2 | Relationship        |
| ------ | ------ | ------------------- |
| ice    | cold   | strong relationship |
| steam  | hot    | strong relationship |


###Word Co-occurrence Matrix

- GloVe first constructs a word co-occurrence matrix.

Example corpus:

    python is a programming language
    java is a programming language

Vocabulary:

[python, programming, language, java]

| Word | python | java | programming | language |
| ---- | --- | --- | --- | --- |
| python  | 0   | 0   | 1   | 1   |
| java  | 0   | 0   | 1   | 1   |
| programming  | 1   | 1   | 0   | 1   |
| language  | 1   | 1   | 1   | 0   |


###Mathematical Intuition

- GloVe learns word vectors so that:

      word_vector1 · word_vector2 ≈ log(co-occurrence probability)

- words appearing together frequently have similar embeddings

- rare combinations have weaker relationships

##Training Process of GloVe

1. Collect a large text corpus

2. Build a word co-occurrence matrix

3. Apply matrix factorization

4. Generate dense word vectors

##Applications of GloVe


- Sentiment Analysis

Detecting positive or negative sentiment.

      Example:

      happy ≈ joyful
      sad ≈ unhappy
- Translation

Understanding relationships between words across languages.

    Example:

    dog (English) ≈ perro (Spanish)
- Chatbots

Understanding user queries and generating responses.

- Document Classification

  Used in tasks like:

     -  spam detection

     -  news categorization

     -  topic classification

###Advantages of GloVe

| Advantage                   | Description                  |
| --------------------------- | ---------------------------- |
| Captures global statistics  | Uses entire corpus           |
| Efficient training          | Faster than some deep models |
| Good semantic relationships | Meaningful word vectors      |


##Limitations of GloVe


1. Context Independent

    Same word always has the same vector.

        Example:

        bank (river)
        bank (financial)

        Both get the same embedding.

2. Static Embeddings

    Vectors do not change based on context.

3. Large Corpus Required

    Training GloVe from scratch requires large datasets.

```python
def extract_pdf_text(file_path):

    text = ""

    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()

    return text
```

```python
pdf_text=extract_pdf_text("/content/5. Intro. to ML & Learning Methods (1).pdf")
```

```python
sentences = sent_tokenize(pdf_text)

print("Total sentences:", len(sentences))
```

```python
embeddings = {}

with open("glove.6B.100d.txt", encoding="utf8") as f:
    for line in f:
        values = line.split()

        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")

        embeddings[word] = vector

print("Loaded words:", len(embeddings))
```

```python
def sentence_embedding(sentence):

    words = word_tokenize(sentence.lower())

    vectors = []

    for word in words:

        if word in embeddings:
            vectors.append(embeddings[word])

    if len(vectors) == 0:
        return np.zeros(100)

    return np.mean(vectors, axis=0)
```

```python
sentence_vectors = []

for sentence in sentences:

    vec = sentence_embedding(sentence)

    sentence_vectors.append(vec)

print("Total sentence vectors:", len(sentence_vectors))
```

```python
def semantic_search(query, top_k=5):

    query_vec = sentence_embedding(query)

    similarities = cosine_similarity(
        [query_vec], sentence_vectors
    )[0]

    top_indices = similarities.argsort()[-top_k:][::-1]

    print("\nTop Results:\n")

    for index in top_indices:

        print("Score:", similarities[index])
        print(sentences[index])
        print()
```

```python
semantic_search("Machine Learning enables systems to extract insights from vast datasets, improving their performance through experience")
```
