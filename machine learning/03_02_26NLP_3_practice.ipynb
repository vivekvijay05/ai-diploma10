#Tokenization

```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize

text = (
    "Natural Language Processing (NLP) is a subfield of Artificial Intelligence." +
    " Machine learning and deep learning to process text and speech." +
    " NLP is widely used in applications like chatbots, voice." +
    " NLP faces challenges such as handling ambiguous language, biases." +
    " Analyzing medical records and research papers to assist in diagnosis."
)
text = text.lower()
sentences = sent_tokenize(text)
sentences
```

```python
from nltk.tokenize import word_tokenize
words=word_tokenize(text)
print(words)
```

# stopwords removal

```python
from nltk.corpus import stopwords
nltk.download("stopwords")

stop_words=set(stopwords.words("english"))
print(stopwords)
```

```python
filtered_words=[]
for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)
filtered_words
```

```python
import nltk
nltk.download("wordnet")
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

Lemmatizer = WordNetLemmatizer()

words=(text)

print("Word Lemmatizer (default - noun) : ")
for word in words:
  print(f"{word:15}-> {Lemmatizer.lemmatize(word)}")
```

```python
import numpy as np
```

```python
filtered_words_array = np.array(filtered_words)
print("Filtered words as a NumPy array:")
print(filtered_words_array)
```

#cleaning

```python
import re
stell = " ".join(filtered_words_array)
clean_text=re.sub(r'[^a-zA-Z0-9\s]', '', stell).strip()
clean_text
```

#Bow

```python
from sklearn.feature_extraction.text import CountVectorizer
doc=(text)
doc
```

```python
bow = CountVectorizer()
X_bow = bow.fit_transform([clean_text])
X_bow
```

```python
bow = CountVectorizer()
X_bow = bow.fit_transform([clean_text])
data=X_bow.toarray()
print(data)
```

```python
features=bow.get_feature_names_out()
print(features)
```

```python
import pandas as pd
df=pd.DataFrame(data, columns=[features])
df
```

#TF_IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_ve=TfidfVectorizer()
to_vector=tf_ve.fit_transform([clean_text])
tf_data=to_vector.toarray()
print(tf_data)
```

```python
tf_feature=tf_ve.get_feature_names_out()
print(tf_feature)
```

```python
tf_df=pd.DataFrame(tf_data,columns=[tf_feature])
tf_df
```
