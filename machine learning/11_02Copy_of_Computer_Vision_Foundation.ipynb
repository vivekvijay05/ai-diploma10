```
# FUNDAMENTALS - IMAGES AS NUMBERS

### What is an Image?

For a computer:

> **An image is a matrix (array) of numbers**

Each number represents **pixel intensity or color information**.



### Pixels and Intensity Values

* Pixel = smallest unit of an image
* Pixel values usually range from **0 to 255**

Examples:

* 0 → Black
* 255 → White

Color pixel example (RGB):

```
[255, 0, 0] → Red
[0, 255, 0] → Green
[0, 0, 255] → Blue

### Image as a Matrix

* Grayscale image → 2D matrix (Height × Width)
* Color image → 3D matrix (Height × Width × Channels)

Example:

```
Image shape: (480, 640, 3)
```

This means:

* 480 rows (height)
* 640 columns (width)
* 3 color channels

##  Color Models

### RGB Color Model

* R → Red
* G → Green
* B → Blue

 Important:
OpenCV loads images in **BGR format**, not RGB.

---

### Grayscale

* Single channel image
* Faster processing
* Used in edge detection, face detection

---

### HSV Color Model

HSV stands for:

* H → Hue (color)
* S → Saturation (color strength)
* V → Value (brightness)

📌 Used for:

* Color detection
* Object tracking

## UNDERSTANDING IMAGE REPRESENTATION



WHAT IS AN IMAGE IN OPENCV?

In OpenCV, an image is represented as a NumPy array.
The shape of this array is:

    (height, width, channels)

CHANNELS:
- 1 channel: Grayscale (intensity only)
- 3 channels: Color (B, G, R in OpenCV, NOT RGB!)
- 4 channels: Color with Alpha (transparency)


      COLOR ORDER IN OPENCV: BGR (Blue, Green, Red)


 Create a 3x3 image with 3 color channels
 Structure:
    
    [ [ [B,G,R], [B,G,R], [B,G,R] ],   Row 1
      [ [B,G,R], [B,G,R], [B,G,R] ],   Row 2
      [ [B,G,R], [B,G,R], [B,G,R] ] ]  Row 3


````python
simple_image = np.array([
    # Row 1: Primary colors
    [[255, 0, 0],    # Pixel (0,0): Blue (max B, no G, no R)
     [0, 255, 0],    # Pixel (0,1): Green
     [0, 0, 255]],   # Pixel (0,2): Red

    # Row 2: Secondary colors (mixes)
    [[255, 255, 0],  # Pixel (1,0): Yellow (B+G)
     [255, 0, 255],  # Pixel (1,1): Magenta (B+R)
     [0, 255, 255]], # Pixel (1,2): Cyan (G+R)

    # Row 3: Grayscale
    [[255, 255, 255],  # Pixel (2,0): White (all colors max)
     [128, 128, 128],  # Pixel (2,1): Gray (equal values)
     [0, 0, 0]]        # Pixel (2,2): Black (no color)
], dtype=np.uint8)  # uint8 = unsigned 8-bit integer (0-255)
````

````python
print(f"\n📊 Image Properties:")
print(f"Shape: {simple_image.shape}")
print(f"   → {simple_image.shape[0]} rows (height)")
print(f"   → {simple_image.shape[1]} columns (width)")
print(f"   → {simple_image.shape[2]} channels (B,G,R)")
print(f"Data type: {simple_image.dtype}")
print(f"Total elements: {simple_image.size}")
````

## THE CRITICAL BGR vs RGB ISSUE

    OpenCV: BGR format (Blue, Green, Red)
    Most other libraries (Matplotlib, PIL): RGB format

````python
# Create a simple image with pure colors
color_test = np.zeros((200, 600, 3), dtype=np.uint8)

# Draw color patches
# Left: Pure Blue in BGR = (255, 0, 0)
color_test[50:150, 50:150] = [255, 0, 0]  # BGR Blue

# Middle: Pure Green in BGR = (0, 255, 0)
color_test[50:150, 250:350] = [0, 255, 0]  # BGR Green

# Right: Pure Red in BGR = (0, 0, 255)
color_test[50:150, 450:550] = [0, 0, 255]  # BGR Red


cv2_imshow(color_test)
````

````python
plt.figure(figsize=(15, 5))

# Plot 1: What OpenCV thinks (correct)
plt.subplot(1, 3, 1)
# We need to convert BGR to RGB for matplotlib
color_test_rgb = cv2.cvtColor(color_test, cv2.COLOR_BGR2RGB)
plt.imshow(color_test_rgb)
plt.title('Correct Display\n(BGR → RGB converted)')
plt.axis('off')
````

````python
import matplotlib.pyplot as plt

# Plot 2: What happens if we don't convert (WRONG!)
plt.subplot(1, 3, 2)
plt.imshow(color_test)  # Displaying BGR as RGB
plt.title('WRONG Display\n(BGR shown as RGB)')
plt.axis('off')
````

````python
# Plot 3: Manual channel swapping
plt.subplot(1, 3, 3)
# Manual conversion: split channels and reorder
b, g, r = cv2.split(color_test)
manual_rgb = cv2.merge([r, g, b])  # BGR to RGB
plt.imshow(manual_rgb)
plt.title('Manual Conversion\n(split → reorder → merge)')
plt.axis('off')

plt.tight_layout()
plt.savefig('Output_Day1/bgr_vs_rgb_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
````

RESIZING METHODS:

cv2.resize(image, (new_width, new_height), interpolation)

    Common interpolation methods:
    - INTER_NEAREST: Fast, pixelated (good for pixel art)
    - INTER_LINEAR: Default, good quality/fast balance
    - INTER_CUBIC: Higher quality, slower
    - INTER_AREA: Good for shrinking (preserves edges)

````python
# Method 1: Specify exact dimensions
resized_exact = cv2.resize(img, (300, 200))  # width, height
cv2_imshow(resized_exact)
````

````python
# Method 2: Scale by factor
scale_factor = 0.5  # Half size
new_width = int(width * scale_factor)
new_height = int(height * scale_factor)
resized_scaled = cv2.resize(img, (new_width, new_height))
print(f"Scaled by 0.5: {new_width}x{new_height}")

cv2_imshow(resized_scaled)
````

````python
# Method 3: Different interpolation methods
print("\nComparing interpolation methods:")
methods = {
    'NEAREST': cv2.INTER_NEAREST,
    'LINEAR': cv2.INTER_LINEAR,
    'CUBIC': cv2.INTER_CUBIC,
    'AREA': cv2.INTER_AREA
}


# Create comparison figure
plt.figure(figsize=(12, 8))
for i, (name, method) in enumerate(methods.items(), 1):
    resized = cv2.resize(img, (200, 133), interpolation=method)
    plt.subplot(2, 2, i)
    plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
    plt.title(f'{name} Interpolation')
    plt.axis('off')

plt.tight_layout()
plt.savefig('Output_Day1/resize_methods_comparison.png', dpi=150)
plt.show()
````

## ROTATING IMAGES

````python
# Get image center
center = (width // 2, height // 2)

# Create rotation matrix
# cv2.getRotationMatrix2D(center, angle, scale)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 degrees

# Apply rotation
# cv2.warpAffine(image, transformation_matrix, output_size)
rotated = cv2.warpAffine(img, rotation_matrix, (width, height))

cv2.imwrite('Output_Day1/rotated_45.jpg', rotated)

cv2_imshow(rotated)
````

## PIXEL-LEVEL MANIPULATION

WORKING WITH INDIVIDUAL PIXELS

````python
# Method 1: Direct pixel access
pixel_value = img[100, 100]  # Gets pixel at (y=100, x=100)
print(f"Pixel at (100, 100): B={pixel_value[0]}, G={pixel_value[1]}, R={pixel_value[2]}")
````

````python
# Method 2: Modify a single pixel
img_copy = img.copy()  # Always work on copies!
img_copy[200, 200] = [0, 255, 255]  # Make it pure red
print(f"Modified pixel to: {img_copy[100, 100]}")

cv2_imshow(img_copy)
````

````python
# Method 3: Modify a region (draw a rectangle)
img_with_box = img.copy()
# Draw a red rectangle from (50,50) to (150,150)
img_with_box[50:150, 50:150] = [0, 0, 255]  # All pixels in this region become red
cv2_imshow(img_with_box)
````

````python
# Method 4: More sophisticated - increase brightness in a region
img_roi_example = img.copy()
# Select Region of Interest (ROI)
roi = img_roi_example[100:300, 100:200]  # Copy of the region

cv2_imshow(roi)
````

````python

# Increase brightness by 50 (clips at 255)
roi_brighter = cv2.add(roi, 50)

cv2_imshow(roi_brighter)
````

````python
# Put it back
img_roi_example[100:300, 100:200] = roi_brighter

cv2_imshow(img_roi_example)
````

## SPLITTING AND MERGING CHANNELS

CHANNEL OPERATIONS:

    Sometimes you need to work on individual color channels.
    Examples:
    - Enhance just the red channel
    - Create special effects
    - Extract color information

````python
# Split into individual channels
b, g, r = cv2.split(img)

print(f"Blue channel shape: {b.shape}")
print(f"Green channel shape: {g.shape}")
print(f"Red channel shape: {r.shape}")

# Display individual channels
plt.figure(figsize=(12, 8))

# Original
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original (RGB)')
plt.axis('off')

# Blue channel (as grayscale)
plt.subplot(2, 3, 2)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

# Green channel
plt.subplot(2, 3, 3)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

# Red channel
plt.subplot(2, 3, 4)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')
````

````python
# Create a false color image (swap channels)
false_color = cv2.merge([r, b, g])  # R as B, B as G, G as R
plt.subplot(2, 3, 5)
plt.imshow(false_color)
plt.title('False Color (R,B,G)')
plt.axis('off')

# Merge back (original order)
merged_back = cv2.merge([b, g, r])
plt.subplot(2, 3, 6)
plt.imshow(cv2.cvtColor(merged_back, cv2.COLOR_BGR2RGB))
plt.title('Merged Back (B,G,R)')
plt.axis('off')

plt.tight_layout()
plt.savefig('Output_Day1/channel_operations.png', dpi=150)
plt.show()
````
