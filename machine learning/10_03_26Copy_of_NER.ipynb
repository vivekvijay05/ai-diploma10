##Named Entity Recognition (NER)

Named Entity Recognition (NER) is a task in Natural Language Processing (NLP) that identifies and classifies important entities in text into predefined categories.

Entities typically include:

- Person names

- Organizations

- Locations

- Dates

- Monetary values

- Products

NER helps machines understand important information in text.


    Elon Musk founded Tesla in California in 2003.

| Word       | Entity Type  |
| ---------- | ------------ |
| Elon Musk  | Person       |
| Tesla      | Organization |
| California | Location     |
| 2003       | Date         |


##Why Named Entity Recognition is Important

NER helps extract meaningful information from large text datasets.

Applications include:

- information extraction

- question answering systems

- chatbots

- search engines

- document classification

      Apple released the iPhone in 2007.

| Entity | Type         |
| ------ | ------------ |
| Apple  | Organization |
| iPhone | Product      |
| 2007   | Date         |

###NER Workflow

    The typical NER process follows these steps:

    Input Text
        ↓
    Text Preprocessing
        ↓
    Tokenization
        ↓
    POS Tagging
        ↓
    NER Model
        ↓
    Entity Classification




Example:

      Input text:

      Barack Obama was born in Hawaii.


Output:


| Word         | Entity   |
| ------------ | -------- |
| Barack Obama | Person   |
| Hawaii       | Location |


##NER Techniques
NER systems can be implemented using different approaches.



1. Rule-Based NER

Uses predefined linguistic rules.

    Example rule:

          Capitalized words → possible entity

    Example:

          New York

          Rules detect it as a location.

Advantages:

- simple

- easy to implement

Disadvantages:

- not flexible

- difficult to scale

2. Machine Learning-Based NER

Uses statistical models trained on labeled data.

Common algorithms:

- Conditional Random Fields (CRF)

- Hidden Markov Models (HMM)

- Support Vector Machines (SVM)

These models learn patterns from training datasets.

3. Deep Learning-Based NER

Modern NER systems use deep learning.

Popular architectures:

- RNN

- LSTM

- BiLSTM-CRF

- Transformer models

Examples include models like BERT.

```python


text = "Apple is looking at buying U.K. startup for $1 billion."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
