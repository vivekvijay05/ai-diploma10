##2. **Vector** Operations & Geometry

**Learning objectives:** Perform vector arithmetic, compute dot/cross product, norms, angle, and visualize in 2D/3D.

```python
# Vector arithmetic
v1 = np.array([2,3])
v2 = np.array([1,4])
print('v1:', v1, '\nv2:', v2)
print('v1 + v2 =', v1+v2)
print('v1 - v2 =', v1-v2)
print('2*v1 =', 2*v1)
```

```python
# vector visualisation

orgin = np.array([0,0])
plt.figure(figsize=(4,3))
plt.quiver(*orgin,*v1,angles="xy",scale_units="xy",scale=1)
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid(True)
plt.show
```

```python
# Dot product and angle
dot = np.dot(v1, v2)
mag_v1 = np.linalg.norm(v1)
mag_v2 = np.linalg.norm(v2)
cos_theta = dot / (mag_v1 * mag_v2)
theta_deg = np.degrees(np.arccos(np.clip(cos_theta, -1.0, 1.0)))
print('Dot:', dot)
print('Magnitudes:', mag_v1, mag_v2)
print('Angle (deg):', theta_deg)
```

## 3. Matrix Operations & Linear Transformations

**Learning objectives:** Matrix addition, multiplication, transpose, identity, matrix-vector multiplication, and geometric interpretation of transformations.

```python
# Matrix basics
M1 = np.array([[1,2],[3,4]])
M2 = np.array([[2,0],[1,2]])
print('M1:\n', M1)
print('\nM2:\n', M2)
print('\nM1 + M2:\n', M1 + M2)
print('\nM1 @ M2:\n', M1 @ M2)
print('\nTranspose M1^T:\n', M1.T)
```

demo practice

```python
x1=np.array([[2,4],[3,4]])
x2=np.array([[4,5],[2,4]])
print('x1:\n', x1)
print('\nx2:\n', x2)
print('\nx1 + x2:\n', x1 + x2)
print('\nx1 @ x2:\n', x1 @ x2)
print('\nTranspose x1^T:\n', x1.T)
```

### Matrix-vector multiplication and interpretation

```python
# Matrix-vector multiplication
M = np.array([[2,0],[0,1]])
x = np.array([3,1])
print('M:\n', M)
print('x:', x)
print('M @ x =', M @ x)  # scales x by 2 in x-direction

# Visualize transformation of unit square by M
square = np.array([[0,0],[1,0],[1,1],[0,1]]).T  # 2x4
trans_sq = M @ square

print(square)
print('\nTransformed square points:\n', trans_sq)
```

```python
# Plot original and transformed square
orig = square.T
trans = trans_sq.T

plt.figure(figsize=(5,5))
# original square
plt.plot(np.append(orig[:,0], orig[0,0]), np.append(orig[:,1], orig[0,1]), '-o', label='Original')
# transformed square
plt.plot(np.append(trans[:,0], trans[0,0]), np.append(trans[:,1], trans[0,1]), '-o', label='Transformed')
plt.axis('equal'); plt.grid(True); plt.legend(); plt.title('Linear Transformation of Unit Square'); plt.show()
```

### Special matrices: Identity, Diagonal, Symmetric

```python
I = np.eye(3)
D = np.diag([2,3,4])
S = np.array([[1,2,3],[2,5,6],[3,6,9]])  # symmetric
print('Identity I:\n', I)
print('\nDiagonal D:\n', D)
print('\nIs S symmetric?', np.allclose(S, S.T))
```

##4. Determinants, Matrix Inverse & Solving Linear Systems

**Learning objectives:** Compute determinant, inverse, understand singular vs non-singular, solve Ax=b using inverse and `np.linalg.solve`, Gaussian elimination concept, and condition number.

```python
# Determinant and inverse
A = np.array([[4,7],[2,6]])
detA = np.linalg.det(A)
print('A:\n', A)
print('\ndet(A):', detA)

if not np.isclose(detA, 0):
    invA = np.linalg.inv(A)
    print('\ninv(A):\n', invA)
    print('\nA @ inv(A):\n', A @ invA)
else:
    print('Matrix is singular; inverse does not exist.')
```

### Solving linear systems Ax = b

```python
# Solve Ax = b
A = np.array([[3,1],[1,2]])
b = np.array([9,8])
x_inv = np.linalg.inv(A) @ b      # using inverse (not recommended for numeric stability)
x_solve = np.linalg.solve(A, b)   # recommended
print('Solution via inverse:', x_inv)
print('Solution via np.linalg.solve:', x_solve)
```

### Rank, singular matrices, and condition number

```python
# Rank and singular example
B = np.array([[1,2],[2,4]])  # second row is multiple of first -> rank 1
print('Rank of B:', np.linalg.matrix_rank(B))
print('Determinant of B:', np.linalg.det(B))

# Condition number
C = np.array([[1,2],[2.0001,4]])
print('Determinant of C:', np.linalg.det(C))
print('\ncond(B):', np.linalg.cond(B))
print('cond(C):', np.linalg.cond(C))  # high cond -> nearly singular
```

### Gaussian elimination — demonstration (simple)

```python
# Simple Gaussian elimination using numpy's solve for a 3x3 example
A = np.array([[2.,1.,-1.],
              [-3.,-1.,2.],
              [-2.,1.,2.]])
b = np.array([8.,-11.,-3.])

# Solve using numpy
x = np.linalg.solve(A,b)
print('Solution x:', x)

# Show that A@x == b (within tolerance)
print('\nA @ x =', A @ x)
print('b =', b)
```
