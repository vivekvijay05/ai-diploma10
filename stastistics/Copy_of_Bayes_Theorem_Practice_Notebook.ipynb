# Bayes' Theorem — Practice

## Q1 — Medical Test (Classic)
In a city, **1%** of people have a disease. A test for the disease returns **positive** with probability **0.99** if the person has the disease (sensitivity), and **positive** with probability **0.05** if the person does not have the disease (false positive rate).

**Question:** If a person tests positive, what is the probability they actually have the disease?

```python
# Q1 — Analytical + Simulation
p_d = 0.01
p_pos_given_d = 0.99
p_pos_given_notd = 0.05

# Analytical (Bayes formula)
p_pos = p_pos_given_d * p_d + p_pos_given_notd * (1 - p_d)
p_d_given_pos = (p_pos_given_d * p_d) / p_pos

print("Analytical: P(D|+)= {:.6f}".format(p_d_given_pos))

# Monte Carlo simulation
import random
trials = 200_000
count_pos = 0
count_d_and_pos = 0
for _ in range(trials):
    has_d = random.random() < p_d
    test_pos = random.random() < (p_pos_given_d if has_d else p_pos_given_notd)
    if test_pos:
        count_pos += 1
        if has_d:
            count_d_and_pos += 1
sim_p = count_d_and_pos / count_pos if count_pos else float('nan')
print("Simulation:  P(D|+)= {:.6f}".format(sim_p))
```

## Q2 — Factory Machines
Three machines A, B, C produce items with proportions and defect rates:
- A: 20% of production, 1% defective
- B: 50% of production, 2% defective
- C: 30% of production, 3% defective

**Question:** If an item is found defective, what is the probability it came from machine C?

```python
# Q2 — Analytical + Simulation
pA, pB, pC = 0.2, 0.5, 0.3
pD_given_A, pD_given_B, pD_given_C = 0.01, 0.02, 0.03

# Analytical
pD = pA*pD_given_A + pB*pD_given_B + pC*pD_given_C
pC_given_D = (pC * pD_given_C) / pD
print("Analytical: P(C|D)= {:.6f}".format(pC_given_D))

# Simulation
import random
trials = 200_000
count_D = 0
count_C_and_D = 0
for _ in range(trials):
    r = random.random()
    if r < pA:
        machine = 'A'
        defective = random.random() < pD_given_A
    elif r < pA + pB:
        machine = 'B'
        defective = random.random() < pD_given_B
    else:
        machine = 'C'
        defective = random.random() < pD_given_C
    if defective:
        count_D += 1
        if machine == 'C':
            count_C_and_D += 1
sim_p = count_C_and_D / count_D if count_D else float('nan')
print("Simulation:  P(C|D)= {:.6f}".format(sim_p))
```

## Q3 — Student & Exam
A student studies with probability 0.7. If he studies, he passes with probability 0.9. If he doesn't study, he passes with probability 0.2.

**Question:** If he passed, what is the probability he had studied?

```python
# Q3 — Analytical + Simulation
pS = 0.7
pPass_given_S = 0.9
pPass_given_notS = 0.2

# Analytical
pPass = pS * pPass_given_S + (1-pS) * pPass_given_notS
pS_given_Pass = (pPass_given_S * pS) / pPass
print("Analytical: P(S|Pass)= {:.6f}".format(pS_given_Pass))

# Simulation
import random
trials = 200_000
count_Pass = 0
count_S_and_Pass = 0
for _ in range(trials):
    studied = random.random() < pS
    passed = random.random() < (pPass_given_S if studied else pPass_given_notS)
    if passed:
        count_Pass += 1
        if studied:
            count_S_and_Pass += 1
sim_p = count_S_and_Pass / count_Pass if count_Pass else float('nan')
print("Simulation:  P(S|Pass)= {:.6f}".format(sim_p))
```

## Q4 — Weather & Wet Ground
P(Rain) = 0.6. If it rains, ground is wet with P=0.7. If it doesn't rain, ground is wet with P=0.1 (sprinklers).

**Question:** If the ground is wet, what is the probability it actually rained?

```python
# Q4 — Analytical + Simulation
pR = 0.6
pW_given_R = 0.7
pW_given_notR = 0.1

# Analytical
pW = pR*pW_given_R + (1-pR)*pW_given_notR
pR_given_W = (pW_given_R * pR) / pW
print("Analytical: P(R|W)= {:.6f}".format(pR_given_W))

# Simulation
import random
trials = 200_000
count_W = 0
count_R_and_W = 0
for _ in range(trials):
    rain = random.random() < pR
    wet = random.random() < (pW_given_R if rain else pW_given_notR)
    if wet:
        count_W += 1
        if rain:
            count_R_and_W += 1
sim_p = count_R_and_W / count_W if count_W else float('nan')
print("Simulation:  P(R|W)= {:.6f}".format(sim_p))
```

## Q5 — Email Spam Filter
30% of emails are spam. The filter flags spam with P=0.99, and flags non-spam incorrectly with P=0.02.

**Question:** If an email is flagged as spam, what is the probability it is actually spam?

```python
# Q5 — Analytical + Simulation
pSpam = 0.3
pFlag_given_S = 0.99
pFlag_given_notS = 0.02

# Analytical
pFlag = pSpam*pFlag_given_S + (1-pSpam)*pFlag_given_notS
pS_given_Flag = (pFlag_given_S * pSpam) / pFlag
print("Analytical: P(Spam|Flag)= {:.6f}".format(pS_given_Flag))

# Simulation
import random
trials = 200_000
count_Flag = 0
count_S_and_Flag = 0
for _ in range(trials):
    is_spam = random.random() < pSpam
    flagged = random.random() < (pFlag_given_S if is_spam else pFlag_given_notS)
    if flagged:
        count_Flag += 1
        if is_spam:
            count_S_and_Flag += 1
sim_p = count_S_and_Flag / count_Flag if count_Flag else float('nan')
print("Simulation:  P(Spam|Flag)= {:.6f}".format(sim_p))
```

## Q6 — Coins in a Box
A box contains two coins: Coin A (fair, P(H)=0.5) and Coin B (biased, P(H)=0.75). A coin is picked uniformly at random and tossed once. If the toss results in Head, what is the probability the chosen coin was Coin B?

```python
# Q6 — Analytical + Simulation
pA = 0.5
pB = 0.5
pH_given_A = 0.5
pH_given_B = 0.75

# Analytical
pH = pA*pH_given_A + pB*pH_given_B
pB_given_H = (pB * pH_given_B) / pH
print("Analytical: P(B|H)= {:.6f}".format(pB_given_H))

# Simulation
import random
trials = 200_000
count_H = 0
count_B_and_H = 0
for _ in range(trials):
    pick_B = random.random() < pB
    head = random.random() < (pH_given_B if pick_B else pH_given_A)
    if head:
        count_H += 1
        if pick_B:
            count_B_and_H += 1
sim_p = count_B_and_H / count_H if count_H else float('nan')
print("Simulation:  P(B|H)= {:.6f}".format(sim_p))
```
