# Python Arrays – Creation, Operations, and Applications


```python
# Import necessary libraries
import numpy as np
import pandas as pd
import array as arr
import sys
import matplotlib.pyplot as plt
```

## 1. Introduction to Arrays

**Notes:**
- **What is an array?** A contiguous block of memory storing elements of the same data type.
- **Why use arrays?**
  - Efficient storage and retrieval compared to other data structures.
  - Ideal for numerical computations and large datasets.
  - Provides a foundation for operations in scientific computing and machine learning.
- Arrays in Python are implemented via different structures: built-in lists, the `array` module, and NumPy arrays.

## 2. Python's Array-like Structures

**Notes:**
- **Lists:**
  - Built-in, flexible data structure that can store elements of different data types.
  - Pros: Easy to use, versatile.
  - Cons: Less memory-efficient for numerical data, slower for large-scale computations.
- **array Module:**
  - More memory-efficient than lists for homogeneous data (same data type).
  - Designed for numeric data, supports specific type codes (e.g., 'i' for integers).
  - Limited functionality compared to NumPy.
- **NumPy Arrays:**
  - Most popular for scientific computing and AI.
  - Advantages:
    - High performance and memory efficiency.
    - Vectorized operations (fast element-wise computations).
    - Broadcasting for operations on arrays of different shapes.
    - Rich library of mathematical functions.

## 3. Creation and Manipulation of Arrays

**Notes:**
- Arrays can be created using lists, the `array` module, or NumPy.
- Accessing elements: Use indexing (e.g., `my_list[0]`, `my_np_array[0]`).
- Modifying elements: Assign new values via indexing.
- Slicing: Extract subarrays using `start:end` syntax.
- Below, we create arrays using different methods and demonstrate memory usage.

```python
my_list=[2,5,4,6,7]
print(my_list)

my_array=arr.array("i",[2,5,4,6,7])
print(my_array)
```

```python
# Creating arrays
# List
my_list = [1, 2, 3, 4, 5]
print("Python List:", my_list)

# array module (type code 'i' for integers)
my_array = arr.array('i', [1, 2, 3, 4, 5])
print("array Module Array:", my_array)

# NumPy array
my_np_array = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", my_np_array)

# Creating arrays with different data types
my_list_mixed = [1, "two", 3.0]  # Mixed types in list
my_array_float = arr.array('f', [1.0, 2.5, 3.7])  # Float array
my_np_array_float = np.array([1.0, 2.5, 3.7], dtype=np.float32)  # NumPy float32
print("\nMixed List:", my_list_mixed)
print("array Module Float Array:", my_array_float)
print("NumPy Float Array:", my_np_array_float)

# Accessing elements
print("\nAccessing first element:")
print("List:", my_list[0])
print("array Module:", my_array[0])
print("NumPy Array:", my_np_array[0])

# Modifying elements
my_list[0] = 10
my_array[0] = 10
my_np_array[0] = 10
print("\nAfter modifying first element:")
print("List:", my_list)
print("array Module:", my_array)
print("NumPy Array:", my_np_array)

# Slicing
print("\nSlicing (elements 1 to 3):")
print("List Slice:", my_list[1:4])
print("array Module Slice:", my_array[1:4])
print("NumPy Array Slice:", my_np_array[1:4])

```

```python
# Memory usage comparison
print("\nMemory Usage:")
print("List:", sys.getsizeof(my_list) + sum(sys.getsizeof(i) for i in my_list), "bytes")
print("array Module:", sys.getsizeof(my_array) + my_array.itemsize * len(my_array), "bytes")
print("NumPy Array:", my_np_array.nbytes + sys.getsizeof(my_np_array), "bytes")
```

## 4. Common Array Operations

**Notes:**
- **Insertion:** Add elements to an array (e.g., `append()`, `np.insert()`).
- **Deletion:** Remove elements by value or index (e.g., `remove()`, `np.delete()`).
- **Searching:** Find the index of an element (e.g., `index()`, `np.where()`).
- **Mathematical Operations:** NumPy supports element-wise operations (e.g., addition, multiplication) and vectorized computations for efficiency.
- Below, we perform these operations on a NumPy array for efficiency and clarity.

```python
n=[8,9]

n.insert(1,90)
n.remove(8)
print(n)
```

```python
[]
# Create a NumPy array for operations
np_array = np.array([10, 20, 30, 40, 50])
print("Original NumPy Array:", np_array)

# Insertion
np_array = np.insert(np_array, 2, 25)  # Insert 25 at index 2
print("\nAfter inserting 25 at index 2:", np_array)

# Deletion
np_array = np.delete(np_array, 3)  # Delete element at index 3
print("After deleting element at index 3:", np_array)



```

```python
# Searching
index = np.where(np_array == 20)[0][0]  # Find index of 20
print("Index of element 20:", index)
```

```python

# Mathematical Operations
np_array2 = np.array([1, 2, 3, 4, 5])
print("\nSecond NumPy Array:", np_array2)
print("Element-wise Addition:", np_array + np_array2)
print("Element-wise Multiplication:", np_array * np_array2)
```

## 5. Applications of Arrays in AI

- **Machine Learning Models:** Training/testing data is typically stored in NumPy arrays for compatibility with libraries like scikit-learn.
- **Vectors and Matrices:** Arrays represent vectors and matrices, the foundation of linear algebra in machine learning.
