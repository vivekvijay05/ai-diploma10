
# 🚀 Functions in Python

## 1. Introduction to Functions

A **function** is like a coffee machine ☕ — you give it ingredients (**input**), it does the work, and gives you coffee (**output**).

**Why use functions?**
- Avoid repeating code (DRY: Don't Repeat Yourself)
- Improve readability
- Make debugging easier



## 2. Types of Functions in Python

### 🔹 Built-in Functions
Examples: `print()`, `len()`, `sum()`, `max()`

### 🔹 User-Defined Functions
Functions you create yourself for specific tasks.


## 4. Parameters and Arguments

### 🔹 Positional arguments – Order matters.
### 🔹 Keyword arguments – Specify by name.
### 🔹 Default arguments – Provide a default value.
### 🔹 Variable-length arguments – *args and **kwargs.


````python
def describe_pet(pet_name, animal_type="dog"):
    print(f"I have a {animal_type} named {pet_name}.")

# Positional arguments
describe_pet("Buddy", "dog")

# Keyword arguments
describe_pet(pet_name="Whiskers", animal_type="cat")

# Default argument
describe_pet("Rocky")
````

````python
def student_details(location,name="achu",course="ba"):
  print(f"i am {name} from {location} studying {course} ")
 #positional arguments
student_details("vivek","diploma in ai","mavelikara")

#keyword arguments
student_details(name="vivek",location="mavelikara",course="diploma in ai")

#default arguments
student_details("mavelikara")
````


## 6. Scope of Variables

- **Local variables**: Declared inside a function, accessible only there.
- **Global variables**: Declared outside functions, accessible everywhere.


````python

x = 10  # Global variable

def local_example():
    y = 5  # Local variable
    print("Inside function, y =", y)

local_example()
print("Outside function, x =", x)
````

Use *args to find the maximum of multiple numbers.

````python
def maximum(*args):
   return max(args)
def minimum(*args):
  return min(args)
def add(*args):
  return sum(args)

null=[]
lim=int(input("ENTER THE LIMIT \n"))

for i in range(1,lim+1):
  y=int(input("ENTER THE NUMBER \n"))
  null.append(y)

print(f"maximum number of {lim} list",{maximum(*null)})

print(f"minimum number of {lim} list",{minimum(*null)})

print(f"sum number of {lim} list",{add(*null)})
````

Factorial using recursion


````python
def fact():
  f=1
  num=int(input("enter the value\n:"))
  a=num
  if num==0:
    print(f"factorial of the number {a} is: 0")
  else:

    for i in range(num):
      f=f*num
      num-=1
    print(f"factorial of the number {a} is:{f}")

fact()
````

Factorial using recursion

````python
def fun(n):
  if n == 0:
    return 1
  elif n < 0:
    print("Factorial is not defined for negative numbers")
    return None
  else:
    return n * fun(n-1)

num = int(input("enter the value\n:"))
result = fun(num)
if result is not None:
  print(f"factorial of the number {num} is: {result}")
````
