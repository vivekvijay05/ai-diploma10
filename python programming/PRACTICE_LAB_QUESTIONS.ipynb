12.Temperature Converter Ask the user for temperature in Celsius, convert it to Fahrenheit using:

Fahrenheit = (Celsius * 9/5) + 32
Print the result.

````python
Celsius=int(input("Enter temperature in celsius\t "))
Fahr=int((Celsius*9/5)+32)
print("Fahrenheit:", Fahr)
Fahren=int(input("Enter temperature in Fahrenheit\t "))
cel=int(5/9*(Fahren-32))
print("Fahrenheit:", cel)
````
