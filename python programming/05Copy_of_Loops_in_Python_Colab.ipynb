## 1️⃣ Introduction to Loops
**Why loops?** To repeat actions without writing the same code again and again.  

**Real-life analogy:** Washing 10 plates — instead of saying *“wash plate 1, wash plate 2…”*, we say *“wash each plate until all are clean.”*



## 2️⃣ For Loop — Basics
Use `for` when you know the number of repetitions or are iterating through a collection (list, tuple, string, range).


````python
# Print numbers from 1 to 5
for i in range(1, 6):
    print(i)
````

````python
# Looping through a list
fruits = ["apple", "banana", "mango"]   #f= fruits[0]='apple'  #f= fruits[1]='banana' #f= fruits[2]='mango'
for f in fruits:
    print(f, end=" ")
````

````python
# Looping through a string
for ch in "PYTHON":
    print(ch, end=" ")
````

````python
birds=["parrot","crow","pheonix","peacock"]
for f in birds:
  print("hello",f)
````

````python
Cart=[]
l=int(input("ENTER THE LIMIT OF CART\n"))
for r in range(l):
   if r==3:
    print("INSIDE IF\n")
add=int(input("ENTER THE CART LIST\n"))
Cart.append(add)

print(f"YOUR CART HISTORY IS\n{Cart}")

lim=int(input("HOW MUCH YOU WANT TO REMOVE\n"))
for r in range(lim):

  delete=int(input("ENTER YOU WANT TO REMOVE\n"))
  if delete in Cart:
    Cart.remove(delete)
    print(f"AFTER REMOVING YOUR CART\n{Cart}")
````


## 3️⃣ While Loop — Basics
Use `while` when the number of repetitions depends on a **condition**.


````python
#coundown from 1 to 5
num=1
while num<6:
  print(num)
  num+=1
````

````python
# @title
# Countdown from 5 to 1
count = 5
while count > 0:
    print(count)
    count -= 1
````


## 4️⃣ Loop Control Statements
- **break** → exit the loop early  
- **continue** → skip the current iteration and continue with the next  
- **pass** → do nothing (placeholder)


````python
# break example — stop when number is found
target = 7
for n in range(1, 11):
    if n == target:
        print(f"Found it: {n}")
        break
    print("Checking:", n)
````

````python
# continue example — print only odd numbers 1..10
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i, end=" ")
````

````python
# pass example — placeholder for future logic
for i in range(3):
    pass  # TODO: implement later
print("\nLoop finished (pass did nothing).")
````


## 5️⃣ Choosing Between Loops
- Use a **for loop** when you **know** how many times to repeat (e.g., iterate over a list of items).  
- Use a **while loop** when repetition depends on a **condition** (e.g., keep asking for input until it's valid).  

**Quick Comparison**  
| Situation | Best Choice | Why |
|---|---|---|
| Processing items in a list | `for` | Fixed, known number of iterations |
| Retry until user enters valid data | `while` | Condition-based repetition |
| Iterate fixed range (1..N) | `for` | Clear, concise |
| Wait for a sensor threshold | `while` | Runs until condition changes |
