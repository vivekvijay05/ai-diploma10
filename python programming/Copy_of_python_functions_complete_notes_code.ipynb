## 2. Parameter Handling

**Notes:**
- **Positional arguments:** passed in order.
- **Keyword arguments:** passed by name; order not important.
- **Default arguments:** provide defaults for parameters.
- **Mutable default caution:** avoid using mutable objects (like list) as default values because they persist across calls.

**Examples & pitfalls**

```python

# Positional and keyword arguments
def introduce(name, age, city="Kottayam"):
    return f"{name}, {age} years old, from {city}"

print(introduce("Anandu", 30))
print(introduce("Ravi", 25, city="Kerala"))
print(introduce(age=22, name="Sita"))  # keyword args can change order

# Mutable default pitfall
def append_bad(item, lst=[]):
    lst.append(item)
    return lst

print(append_bad(1))
print(append_bad(2))  # unexpected: previous list reused

# Correct pattern
def append_good(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(append_good(1))
print(append_good(2))
```

## 3. Return Values

**Notes:**
- Use `return` to send value(s) back to caller.
- If no `return`, function returns `None`.
- Can return multiple values (they are packed into a tuple).

**Examples**

```python

def no_return(a):
    a += 1  # modifies local copy
    # no explicit return

def multiple(a, b):
    return a+b, a*b, (a, b)  # returns tuple

print(no_return(5))  # None
print(multiple(3,4))  # (7, 12, (3,4))

# Unpacking multiple returns
sum_, prod, pair = multiple(5,6)
print("sum", sum_, "prod", prod, "pair", pair)
```

## 4. Variable Scope — Local vs Global

**Notes:**
- Variables defined inside a function are **local** to that function.
- Variables defined at module level are **global**.
- To read a global variable inside a function: no keyword needed. To reassign, use `global` keyword (but prefer returning values).
- `nonlocal` used to refer to variable in enclosing (but non-global) scope.

**Examples:**

```python

x = "global_x"

def show():
    x = "local_x"
    print("inside show:", x)

def show_global_read():
    print("reading global x:", x)  # allowed

def modify_global():
    global x
    x = "modified_global_x"

def outer():
    y = "outer_y"
    def inner():
        nonlocal y
        y = "inner_modified_y"
        return y
    print("before inner:", y)
    print("inner returned:", inner())
    print("after inner:", y)

show()
show_global_read()
print("before modify_global:", x)
modify_global()
print("after modify_global:", x)
outer()
```

## 6. Global & Local Variables — Best Practices

**Notes:**
- Prefer function arguments and return values over modifying globals.
- Keep functions pure if possible: given same inputs, return same outputs and cause no side effects.
- Use globals sparingly for constants.

**Example refactor**

```python

# BAD: function modifies global
counter = 0
def inc_bad():
    global counter
    counter += 1

# GOOD: function returns new value
def inc_good(c):
    return c + 1

# Demonstration
counter = 0
inc_bad()
print("counter after inc_bad:", counter)  # changed globally

counter = 0
counter = inc_good(counter)
print("counter after inc_good:", counter)
```

## 7. Recursion

**Notes:**
- Recursion is when a function calls itself.
- Always define a base case to stop recursion.
- Recursive solutions can be elegant but may hit recursion depth or be less efficient.

**Examples: factorial using recursion

```python
# Function to calculate factorial using recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1  # Base case
    return n * factorial(n - 1)  # Recursive case

# Input from user
num = int(input("Enter a positive integer: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print(f"Factorial of {num} is {factorial(num)}")
```

## 9. `self` in Instance Methods (Classes)

**Notes:**
- `self` refers to the instance and is passed automatically when calling instance methods.
- It is just a naming convention; any valid identifier works but use `self` for readability.

**Examples**

```python

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age  # instance attributes
    def info(self):
        return f"{self.name}, {self.age}"

p = Person("Anandu", 30)
print(p.info())

# Demonstrate that method call implicitly passes self
print(Person.info(p))  # equivalent to p.info()
```

## 12. `map`, `filter`, `reduce`

**Notes:**
- `map(func, iterable)`: apply func to each element; returns iterator.
- `filter(func, iterable)`: keep elements where func(elem) is True.
- `reduce(func, iterable)`: combine elements using binary func; import from `functools`.

**Examples & when to prefer list comprehensions**

```python

from functools import reduce

nums = [1,2,3,4,5]

mapped = list(map(lambda x: x*2, nums))
filtered = list(filter(lambda x: x%2==1, nums))
reduced = reduce(lambda a,b: a+b, nums)

print("map:", mapped)
print("filter:", filtered)
print("reduce:", reduced)

# Equivalent list comprehensions (often clearer)
print("comprehension map:", [x*2 for x in nums])
print("comprehension filter:", [x for x in nums if x%2==1])
```

## 13. Inner Functions & Closures

**Notes:**
- Inner functions are functions defined inside other functions.
- Closures capture variables from the enclosing scope, allowing state to persist.
- Use `nonlocal` to modify enclosing variables.

**Examples**

```python
count = 0
def make_counter():

    def inc():
        global count
        count += 1
        return count
    return inc

c1 = make_counter()
print(c1(), c1(), c1())  # 1 2 3
```

```python


# Closure example with parameters
def make_multiplier(n):
    def mul(x):
        return x * n
    return mul

double = make_multiplier(2)
print(double(5))
print(double(9))
```

## 14. Decorators

**Notes:**
- Decorators wrap a function to extend behavior without modifying the function body.
- Common use-cases: logging, timing, access control, caching.
- Use `@wraps` from `functools` to preserve metadata.

**Step-by-step examples:**
1. Simple decorator that prints before/after.
2. Parameterized decorator (decorator factory).
3. Using decorators for caching (simple memoize).


```python

from functools import wraps
import time, functools

# 1. Simple decorator
def announce(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}...")
        result = func(*args, **kwargs)
        print(f"{func.__name__} finished.")
        return result
    return wrapper

@announce
def say(msg):
    print(msg)

say("Hello with decorator!")
```

```python


# 2. Parameterized decorator (factory)
def repeat(times=2):
    def deco(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = None
            for _ in range(times):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return deco

@repeat(times=3)
def echo(x):
    print(x)

echo("Echoed")

# 3. Simple memoize decorator
def memoize(func):
    cache = {}
    @wraps(func)
    def wrapper(*args):
        if args in cache:
            return cache[args]
        result = func(*args)
        cache[args] = result
        return result
    return wrapper

@memoize
def slow_fib(n):
    if n < 2:
        return n
    return slow_fib(n-1) + slow_fib(n-2)

print("memoized slow_fib(30):", slow_fib(30))
```

## 15. Mini Project — Utility Library (Combine Topics)

**Requirements:**
- Function `aggregate(*nums, operation='sum')` supporting 'sum', 'prod', 'avg'.
- Use `*args` and `**kwargs` where relevant.
- Add a timing decorator and a input validation decorator.
- Provide both functional and decorator-based implementations.


```python

from functools import reduce, wraps
import time

# timing decorator
def timing(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end-start:.6f}s")
        return res
    return wrapper

# input validation decorator
def validate_nums(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if not args:
            return None
        for v in args:
            if not isinstance(v, (int, float)):
                raise TypeError("All inputs must be numbers")
        return func(*args, **kwargs)
    return wrapper

@timing
@validate_nums
def aggregate(*nums, operation='sum'):
    if operation == 'sum':
        return reduce(lambda a,b: a+b, nums)
    elif operation == 'prod':
        return reduce(lambda a,b: a*b, nums)
    elif operation == 'avg':
        return reduce(lambda a,b: a+b, nums)/len(nums)
    else:
        raise ValueError("Unknown operation")

print("sum:", aggregate(1,2,3,4,5, operation='sum'))
print("prod:", aggregate(1,2,3,4,5, operation='prod'))
print("avg:", aggregate(1,2,3,4,5, operation='avg'))
```
