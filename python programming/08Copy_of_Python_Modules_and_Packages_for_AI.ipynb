**📚 Assignment**


Create a module stats.py that includes:

* Mean

* Median

* Mode

Use that module in another file to analyze a list of student marks.

```python
%%writefile stats.py

med=0
def mean (*args):
  total_sum=sum(args)
  count=len(args)
  x=total_sum/count
  return(x)

def median(*args):

  if len(args)%2==1:
    args=list(args)
    args.sort()
    med=len(args)//2
    med_value=args[med]
    return med_value

  else:
    args=list(args)
    args.sort()
    med=len(args)//2
    med_value=(args[med]+args[med-1])/2
    return med_value


def mode(*args):
  frequency={}
  for a in args:
    if a in frequency:
      frequency[a]+=1
  else:
      frequency[a]=1
      max_frequency=max(frequency.values())
      mode=[key for key,values in frequency.items()if values==max_frequency]

  if len(mode)==len(args):
    print("no mode exist!!")
  else:
    print("mode is",mode)
```

```python
from stats import mean,median,mode
l=[]
n=int(input("enter no of subject\t"))
for s in range(n):
  lis=int(input("enter the marks:::"))
  l.append(lis)
print("mean mark",mean(*l))
print("median mark",median(*l))
mode(*l)
```
