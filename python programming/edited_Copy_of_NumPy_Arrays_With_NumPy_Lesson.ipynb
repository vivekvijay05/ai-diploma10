## 3) Indexing and slicing

Access elements, rows, columns, and slices.

```python
A = np.arange(24).reshape(3,4,2)
print('A:\n', A)
print('\nA[1,2] =', A[1,3,1])  # second row, third column
```

```python

print('\nA[0] (first row) =', A[0])
print('\nA[:,1] (second column) =', A[:,1])
print('\nA[1:, 2:] =\n', A[1:, 2:])

# modifying a slice affects the array (views vs copies)
sub = A[0:2, 1:3]
print('\nsub before:\n', sub)
sub[0,0] = 999
print('\nA after modifying sub:\n', A)

# To avoid this, use copy()
sub_copy = A[0:2, 1:3].copy()
sub_copy[0,0] = -1
print('\nA remains (after modifying copy):\n', A)
```

## 5) Broadcasting

Operations between differently-shaped arrays where shapes are compatible.

```python
M = np.array([[1,2,3],[4,5,6],[5,6,1]])
v = np.array([10,20,30])
print('M:\n', M)
print('\nv:', v)
print('\nM + v (broadcast v across rows) =\n', M + v)


# Broadcasting with column vector
w = np.array([[1],[2],[3]])

print(w)
print('\nM + w =\n', M + w)
```

## 7) Aggregation and statistics

Functions: sum, mean, median, std, max, min (use axis argument).

```python
data = np.random.randn(1000).reshape(250,4)
print('data shape:', data.shape)
print('mean (overall):', data.mean())
print('mean (axis=0):', data.mean(axis=0)) #axis=0: Aggregates down the rows for each column.(to find colomns)axis=1(to find rows)

print('std (axis=1)[:5]:', data.std(axis=1)[:5]) #axis=1: Aggregates across the columns for each row.
print('max per column:', data.max(axis=0))
```

## 8) Performance: Python list vs NumPy array

Vectorized NumPy operations are usually much faster for large numeric data.

```python
import time
import numpy as np
N = 200000
py_list = list(range(N))
np_arr = np.arange(N)

# sum with Python
t0 = time.time()
s1 = sum(py_list)
t1 = time.time()

# sum with NumPy
t2 = time.time()
s2 = np_arr.sum()
t3 = time.time()

print('Python sum time: {:.4f}s'.format(t1-t0))
print('NumPy sum time: {:.4f}s'.format(t3-t2))
print('Speedup ~ {:.1f}x'.format((t1-t0)/(t3-t2)))
```

## 9) Visualization (Matplotlib)

Plotting a simple function and an image representation of a 2D array. (Matplotlib defaults used.)

```python
import numpy as np

image = np.array([
    [1000, 120, 130],
    [90, 110, 140],
    [80, 100, 120]
])

mean = image.mean(axis=0)
std = image.std(axis=0)

normalized_image = (image - mean) / std
print(normalized_image)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8,3))
plt.plot(x, y)
plt.title('Sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()

# 2D array visualization
Z = np.arange(100).reshape(10,10)
plt.figure(figsize=(4,4))
plt.imshow(Z, cmap='gray')
plt.title('10x10 grid visualization (imshow)')
plt.colorbar()
plt.show()
```
