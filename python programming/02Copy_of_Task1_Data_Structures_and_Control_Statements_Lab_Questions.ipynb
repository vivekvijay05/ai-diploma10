### Task 11: List and Match Statement
Create a list [1, 2, 3, 4]. Use `match` on the list's length: match len(my_list): case 4: print('Perfect size'); case n if n > 4: print('Too long'); case _: print('Other size'). Then, if the length matches 4, append 5; else remove the first element.

```python
# Create the list
my_list = [1, 2, 3, 4]

# Use match on the list's length
match len(my_list):
    case 4:
        print("Perfect size")
        # If length is 4, append 5
        my_list.append(5)
    case n if n > 4:
        print("Too long")
        # If length > 4, remove the first element
        my_list.pop(0)
    case _:
        print("Other size")
        # For other cases, remove the first element
        my_list.pop(0)

# Print the final list
print(my_list)





```

### Task 14: Dictionary with Match and Nesting
