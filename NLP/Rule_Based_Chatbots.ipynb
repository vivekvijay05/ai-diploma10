Regex-Based Chatbot

```python

import re

patterns = [
    (r"hi|hello|hey", "Hello! 👋"),
    (r"how are you", "I am fine, thank you! 😊"),
    (r"what is your name|your name", "I am a rule-based chatbot."),
    (r"my name is (.*)", "Nice to meet you, {}!"),
    (r"course|courses|ai|ml", "We offer AI, ML, and Data Science courses."),
    (r"fees|fee", "Fees depend on the course. Please contact office."),
    (r"location|where are you", "We are located in Kerala."),
    (r"bye|goodbye", "Goodbye! Have a nice day! 👋")
]

def regex_chatbot(user_input):
    for pattern, response in patterns:
        match = re.search(pattern, user_input.lower())

        if match:

            if "{}" in response:
                return response.format(match.group(1))
            return response

    return "Sorry, I didn't understand that."

print("Regex Chatbot (type 'exit' to stop)\n")

while True:
    user = input("You: ")

    if user.lower() == "exit":
        print("Bot: Goodbye! 👋")
        break

    print("Bot:", regex_chatbot(user))
```

#Chatbot with Memory

```python
import re

memory = {}

def chatbot(user_input):

    # Store name
    match = re.search(r"my name is (.*)", user_input.lower())
    if match:
        memory["name"] = match.group(1)
        return f"Nice to meet you {memory['name']}"

    # Recall name
    if "what is my name" in user_input.lower():
        if "name" in memory:
            return f"Your name is {memory['name']}"
        else:
            return "I don't know your name yet"

    # Store favorite subject
    match = re.search(r"i like (.*)", user_input.lower())
    if match:
        memory["interest"] = match.group(1)
        return f"Great! I will remember you like {memory['interest']}"

    # Recall interest
    if "what do i like" in user_input.lower():
        if "interest" in memory:
            return f"You like {memory['interest']}"
        else:
            return "I don't know yet"

    # Basic responses
    if re.search(r"hi|hello", user_input.lower()):
        return "Hello!"

    if "bye" in user_input.lower():
        return "Goodbye!"

    return "I don't understand"

# Run chatbot
while True:
    user = input("You: ")

    if user == "exit":
        break

    print("Bot:", chatbot(user))
```

NLTK Chatbot

```python


from nltk.chat.util import Chat, reflections


pairs = [
    (r"hi|hello|hey", ["Hello!", "Hi there! 👋"]),
    (r"how are you ?", ["I am fine, thank you!", "Doing great! 😊"]),
    (r"what is your name ?", ["I am an NLTK chatbot."]),
    (r"courses ?", ["We offer AI, ML, Data Science courses."]),
    (r"fees ?", ["Fees depend on the course."]),
    (r"bye", ["Goodbye! 👋", "See you soon!"])
]


chatbot = Chat(pairs, reflections)

print("\nNLTK Chatbot (type 'quit' to stop)\n")


chatbot.converse()
```

NLTK Chatbot with Memory

```python
from nltk.chat.util import Chat, reflections

memory = {}

pairs = [
    (r"hi|hello", ["Hello!"]),
    (r"bye", ["Goodbye!"])
]

chat = Chat(pairs, reflections)

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    # Memory logic]
    if "my name is" in user_input.lower():
        name = user_input.split("my name is")[1].strip()
        memory["name"] = name
        print(f"Bot: Nice to meet you {name}")
        continue

    if "what is my name" in user_input.lower():
        if "name" in memory:
            print(f"Bot: Your name is {memory['name']}")
        else:
            print("Bot: I don't know your name yet")
        continue

    # Default NLTK response
    print("Bot:", chat.respond(user_input))
```
