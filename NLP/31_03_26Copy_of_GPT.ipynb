#Encoder vs Decoder

- Encoder reads the entire input sentence and converts it into a meaningful representation (context vector).

- Decoder generates output one word at a time using previous words.

| Feature       | Encoder          | Decoder               |
| ------------- | ---------------- | --------------------- |
| Purpose       | Understand input | Generate output       |
| Direction     | Bidirectional    | Unidirectional        |
| Input access  | Full sentence    | Only previous words   |
| Output        | Context vector   | Sequence (text)       |
| Attention     | Self-attention   | Masked self-attention |
| Example Model | BERT             | GPT                   |


      Encoder → Understanding
      Decoder → Generating
      Together → Complete AI system

##Key Components of GPT

###1. Tokenization

Text is split into tokens.

    Example:

    "I love NLP"
    → ["I", "love", "NLP"]

###2. Embeddings

    Each token → vector

    love → [0.25, 0.78, ...]

###3. Positional Encoding

    Adds word order:

    Word + Position → Final vector

###4. Masked Self-Attention

GPT can only see previous words, not future words

Example


    The cat sat on the mat

When predicting sat, GPT sees:

    The cat

Cannot see "on the mat"

###5. Feed Forward Network

Applies transformation:

###6. Softmax Output

Converts output into probabilities:

mat → 0.75  
floor → 0.10  
table → 0.15  

    Each word depends on All previous words

## Sampling Techniques


1. Greedy Search

- Always pick highest probability
- Output is predictable


2. Temperature

    Controls randomness:

| Value | Effect        |
| ----- | ------------- |
| 0.2   | deterministic |
| 0.7   | balanced      |
| 1.0   | creative      |


3. Top-k Sampling

        Select from top k words


4. Top-p (Nucleus Sampling)

        Select words with cumulative probability

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")

print("GPT Model Loaded")
```

```python
prompt = "Artificial Intelligence is"

result = generator(prompt, max_length=50)

print(result[0]['generated_text'])
```

```python
#Temperature
result = generator(
    "Once upon a time",
    max_length=50,
    temperature=0.7
)

print(result[0]['generated_text'])
```

```python
#Top-k Sampling

result = generator(
    "The future of AI is",
    max_length=50,
    top_k=50
)

print(result[0]['generated_text'])
```

```python
#Top-p Sampling

result = generator(
    "Machine learning is",
    max_length=50,
    top_p=0.9
)

print(result[0]['generated_text'])
```

```python
prompts = [
    "Explain machine learning in simple terms:",
    "Write a short story about a robot:",
    "Benefits of AI in education:",
]

for p in prompts:
    result = generator(p, max_length=60)
    print("\nPrompt:", p)
    print(result[0]['generated_text'])
```
